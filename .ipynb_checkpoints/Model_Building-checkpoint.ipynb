```python
# Importing all the necessary libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
df = pd.read_csv("Cleaned-House_data.csv")
df.head()
```

```python
# we will be using dummies or one hot encoding methods to convert the text values of location into the numeric variables
df_dum = pd.get_dummies(df['location'])
df_dum.head()
```

```python
# we will concatenate this dataframe into the original one
df_new = pd.concat([df,df_dum.drop('other', axis='columns')], axis='columns')
df_new.head()
```

```python
# let's us drop the location feature after we are done creating dummies
df_new1 = df_new.drop(['location'], axis = 1)
df_new1
```

```python
# Further, we will create our predictors and target variables
X = df_new1.drop('price', axis=1)
X.head()
```

```python
y = df_new1['price']
y.head()
```

```python
# dividing the dataset into 70% training and 30% testing for our linear regression model
np.random.seed(12)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# implementing K-fold cross-validation to make sure we have the optimal model
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.3, random_state=0)
cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
# optimizing our model by implementing GridSearchCV
# we will also compare this linear regression model to two other models (decision trees & lasso regression)
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def best_model(X,y):
    algo = {
        'Linear Regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso':{
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion':['mse','friedman_mse'],
                'splitter':['best','random']
            }
        }
    }
    score = []
    cv = ShuffleSplit(n_splits=5, test_size=0.3, random_state=0)
    for algo_name, a in algo.items():
        gs = GridSearchCV(a['model'], a['params'], cv=cv, return_train_score = False)
        gs.fit(X,y)
        score.append({
            'model':algo_name,
            'best_score':gs.best_score_,
            'best_params':gs.best_params_
        })
        
    return pd.DataFrame(score, columns=['model','best_score','best_params'])
best_model(X,y)
```

```python
# Based on the above results, it can be said that Linear Regression performs better
```

```python
# creating a fucntion to receive some values and predict what can be the average price
def predict_price(location, sqft, bath, bhk):
    loc_index = np.where(X.columns==location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1
        
    return model.predict([x])[0]
```

```python
predict_price('2nd Stage Nagarbhavi',1000,2,2)
```

```python
predict_price('Abbigere',800,1,2)
```

```python
predict_price('Indira Nagar',1000, 3, 3)
```
